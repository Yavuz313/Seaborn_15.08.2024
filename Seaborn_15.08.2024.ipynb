```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
```

## Scatter Plot
- onemli ve cok is kolaylastirir

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips)
plt.show()
```

```python
sns.regplot(data = tips, y = "tip", x = "total_bill");
# Fazla odenen fiyat bahsisi arttiriyor mu bu regresyondur. korelasyon ilişki, regresyon etki
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "sex") # hue icine vercegimiz data katogorik olmali
plt.show()
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "smoker") 
plt.show()
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "day") 
plt.show()
```

## Style

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "smoker", style = "sex") # sigara icenlerin cinsiyetini gosterdi
plt.show()
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "smoker", style = "day")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10 ,5))

sns.scatterplot(x = "total_bill", y = "tip", data = tips, style = "sex", ax = ax[0])
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "smoker", ax = ax[1]);
```

```python
# palette

plt.figure(figsize = (8, 6))

sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "sex", palette = "plasma");
```

```python

plt.figure(figsize = (8, 6))

sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "sex", palette = "cubehelix");
```

```python
plt.figure(figsize = (8, 6))

sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "sex", palette = "cubehelix", alpha = 0.4); # alpha rengin yogunlugunu belirler
```

```python
plt.figure(figsize = (8, 6))

sns.scatterplot(x = "total_bill", 
                y = "tip", 
                data = tips, 
                hue = "sex", 
                palette = "cubehelix", 
                alpha = 0.7,
                s = 150); # icindeki dairelerin size ni belirler
```
